````python
RED = 0
BLACK = 1

class Node:
    def __init__(self, key):
        self.key = key
        self.color = RED
        self.left = None
        self.right = None
        self.parent = None


class RedBlackTree:
    def __init__(self):
        # Sentinel NIL node (berwarna BLACK)
        self.NIL = Node(0)
        self.NIL.color = BLACK
        self.root = self.NIL


    # LEFT ROTATE

    def left_rotate(self, x):
        y = x.right
        x.right = y.left

        if y.left != self.NIL:
            y.left.parent = x

        y.parent = x.parent

        if x.parent is None:
            self.root = y
        elif x == x.parent.left:
            x.parent.left = y
        else:
            x.parent.right = y

        y.left = x
        x.parent = y


    # RIGHT ROTATE

    def right_rotate(self, x):
        y = x.left
        x.left = y.right

        if y.right != self.NIL:
            y.right.parent = x

        y.parent = x.parent

        if x.parent is None:
            self.root = y
        elif x == x.parent.right:
            x.parent.right = y
        else:
            x.parent.left = y

        y.right = x
        x.parent = (y)

    def insert_fixup(self, z):
        while z.parent and z.parent.color == RED:
            if z.parent == z.parent.parent.left:
                y = z.parent.parent.right
                if y.color == RED:

                    z.parent.color = BLACK
                    y.color = BLACK
                    z.parent.parent.color = RED
                    z = z.parent.parent
                else:
                    if z == z.parent.right:
                        z = z.parent
                        self.left_rotate(z)
                    z.parent.color = BLACK
                    z.parent.parent.color = RED
                    self.right_rotate(z.parent.parent)
            else:
                y = z.parent.parent.left
                if y.color == RED:
                    z.parent.color = BLACK
                    y.color = BLACK
                    z.parent.parent.color = RED
                    z = z.parent.parent
                else:
                    if z == z.parent.left:
                        z = z.parent
                        self.right_rotate(z)
                    z.parent.color = BLACK
                    z.parent.parent.color = RED
                    self.left_rotate(z.parent.parent)

        self.root.color = BLACK

    # INSERT
    def insert(self, key):
        z = Node(key)
        z.left = self.NIL
        z.right = self.NIL

        y = None
        x = self.root

        while x != self.NIL:
            y = x
            if z.key < x.key:
                x = x.left
            else:
                x = x.right

        z.parent = y

        if y is None:
            self.root = z
        elif z.key < y.key:
            y.left = z
        else:
            y.right = z

        z.color = RED
        self.insert_fixup(z)


    # SEARCH  ← inti tugas

    def search(self, key):
        x = self.root

        while x != self.NIL and key != x.key:
            if key < x.key:
                x = x.left
            else:
                x = x.right

        if x == self.NIL:
            return None   # tidak ditemukan
        return x          # ditemukan


     # INORDER (untuk verifikasi)
    def inorder(self, node, result=None):
        if result is None:
            result = []
        if node != self.NIL:
            self.inorder(node.left, result)
            warna = "RED" if node.color == RED else "BLACK"
            result.append(f"{node.key}({warna})")
            self.inorder(node.right, result)
        return result



# MAIN - Demo pencarian
if __name__ == "__main__":
    rbt = RedBlackTree()

    data = [10, 20, 30, 15, 25, 5, 1, 7]
    for val in data:
        rbt.insert(val)

    print("Isi pohon (inorder):")
    print(" -> ".join(rbt.inorder(rbt.root)))

    print("\n--- UJI SEARCH ---")
    cari = [15, 25, 99, 1]
    for k in cari:
        hasil = rbt.search(k)
        if hasil:
            warna = "RED" if hasil.color == RED else "BLACK"
            print(f"Cari {k}: DITEMUKAN | warna = {warna}")
        else:
            print(f"Cari {k}: TIDAK DITEMUKAN")
````

## **Analisis Worst-Case dan Average-Case untuk Red-Black Tree**

Red-Black Tree adalah jenis pohon pencarian biner (Binary Search Tree - BST) yang menyeimbangkan diri. Properti utamanya menjamin bahwa tinggi pohon selalu logaritmik terhadap jumlah node (n). Secara spesifik, tinggi maksimum sebuah Red-Black Tree dengan n node adalah sekitar `2 * log2(n+1)`. Karena properti ini, sebagian besar operasi pada Red-Black Tree memiliki kompleksitas waktu logaritmik.

Berikut adalah analisis untuk operasi-operasi utama:

1.  **Pencarian (`search(key)`)**
    *   **Worst-Case Time Complexity: O(log n)**
        *   Untuk mencari sebuah elemen, kita harus melintasi pohon dari akar ke daun. Karena tinggi pohon Red-Black Tree dijamin logaritmik, dalam kasus terburuk (yaitu, mencari elemen yang tidak ada atau elemen yang berada di jalur terpanjang), jumlah perbandingan dan langkah yang dibutuhkan akan sebanding dengan tinggi pohon.
    *   **Average-Case Time Complexity: O(log n)**
        *   Sama seperti worst-case, karena sifatnya yang seimbang, tinggi pohon selalu logaritmik, sehingga rata-rata waktu pencarian juga O(log n).

2.  **Penyisipan (`insert(key)`)**
    *   Operasi penyisipan terdiri dari dua bagian utama:
        a.  **Penyisipan BST standar**: Mencari posisi yang tepat untuk node baru dan menyisipkannya (seperti pada BST biasa).
        b.  **Perbaikan (`insert_fixup`)**: Menyeimbangkan kembali pohon dengan rotasi dan pewarnaan ulang jika aturan Red-Black Tree dilanggar.
    *   **Worst-Case Time Complexity: O(log n)**
        *   Bagian (a) membutuhkan waktu O(log n) karena traversal hingga daun.
        *   Bagian (b) (`insert_fixup`) melibatkan traversal ke atas dari node yang baru disisipkan hingga ke akar (maksimal O(log n) langkah). Di setiap langkah, paling banyak terjadi beberapa pewarnaan ulang (O(1)) dan maksimal dua operasi rotasi (O(1) per rotasi). Oleh karena itu, `insert_fixup` secara keseluruhan juga membutuhkan waktu O(log n).
        *   Gabungan dari kedua bagian ini menghasilkan kompleksitas O(log n).
    *   **Average-Case Time Complexity: O(log n)**
        *   Untuk alasan yang sama dengan worst-case, rata-rata waktu penyisipan juga O(log n).

3.  **Rotasi (`left_rotate(x)` dan `right_rotate(x)`)**
    *   **Worst-Case Time Complexity: O(1)**
        *   Operasi rotasi melibatkan sejumlah tetap manipulasi pointer (mengubah parent, child, dan sibling). Ini tidak tergantung pada jumlah node dalam pohon.
    *   **Average-Case Time Complexity: O(1)**
        *   Sama seperti worst-case, ini adalah operasi waktu konstan.

4.  **Traversal Inorder (`inorder(node, result)`)**
    *   **Worst-Case Time Complexity: O(n)**
        *   Traversal inorder mengunjungi setiap node dalam pohon tepat satu kali. Oleh karena itu, waktu yang dibutuhkan sebanding dengan jumlah total node.
    *   **Average-Case Time Complexity: O(n)**
        *   Sama seperti worst-case, karena setiap node harus dikunjungi.

**Kesimpulan:**

Red-Black Tree adalah struktur data yang sangat efisien untuk operasi pencarian, penyisipan, dan penghapusan karena semua operasi dasar ini dijamin memiliki kompleksitas waktu O(log n), bahkan dalam kasus terburuk. Ini menjadikannya pilihan yang sangat baik untuk aplikasi di mana kinerja yang konsisten penting, seperti dalam implementasi `std::map` atau `TreeMap` di berbagai library standar.
