## 1. Bubble Sort

### Pseudocode

```
FUNGSI bubble_sort(array):
  n = panjang array
  UNTUK i DARI 0 HINGGA n-2:
    UNTUK j DARI 0 HINGGA n-i-2:
      JIKA array[j] > array[j+1]:
        TUKAR array[j] DAN array[j+1]
  KEMBALIKAN array
```

````python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        # Last i elements are already in place
        for j in range(n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Demo
example_list = [64, 34, 25, 12, 22, 11, 90]
print(f"Original list: {example_list}")
sorted_list = bubble_sort(example_list)
print(f"Sorted list (Bubble Sort): {sorted_list}")
````

### Analisis Kompleksitas

*   **Kompleksitas Waktu:**
    *   **Kasus Terburuk:** O(n^2) (misalnya, array yang diurutkan terbalik)
    *   **Kasus Rata-rata:** O(n^2)
    *   **Kasus Terbaik:** O(n) (jika array sudah terurut, dengan optimasi untuk berhenti lebih awal)
*   **Kompleksitas Ruang:** O(1) (pengurutan in-place)

## 2. Selection Sort

### Pseudocode

```
FUNGSI selection_sort(array):
  n = panjang array
  UNTUK i DARI 0 HINGGA n-2:
    min_idx = i
    UNTUK j DARI i+1 HINGGA n-1:
      JIKA array[j] < array[min_idx]:
        min_idx = j
    TUKAR array[min_idx] DAN array[i]
  KEMBALIKAN array
```

````python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        # Find the minimum element in remaining unsorted array
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        # Swap the found minimum element with the first element
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# Demo
example_list = [64, 25, 12, 22, 11]
print(f"Original list: {example_list}")
sorted_list = selection_sort(example_list)
print(f"Sorted list (Selection Sort): {sorted_list}")
````

### Analisis Kompleksitas

*   **Kompleksitas Waktu:**
    *   **Kasus Terburuk:** O(n^2)
    *   **Kasus Rata-rata:** O(n^2)
    *   **Kasus Terbaik:** O(n^2)
*   **Kompleksitas Ruang:** O(1) (pengurutan in-place)

## 3. Insertion Sort

### Pseudocode

```
FUNGSI insertion_sort(array):
  n = panjang array
  UNTUK i DARI 1 HINGGA n-1:
    key = array[i]
    j = i - 1
    SELAMA j >= 0 DAN key < array[j]:
      array[j+1] = array[j]
      j = j - 1
    array[j+1] = key
  KEMBALIKAN array
```

````python
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        # Move elements of arr[0..i-1], that are greater than key,
        # to one position ahead of their current position
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Demo
example_list = [12, 11, 13, 5, 6]
print(f"Original list: {example_list}")
sorted_list = insertion_sort(example_list)
print(f"Sorted list (Insertion Sort): {sorted_list}")
````

### Analisis Kompleksitas

*   **Kompleksitas Waktu:**
    *   **Kasus Terburuk:** O(n^2) (misalnya, array yang diurutkan terbalik)
    *   **Kasus Rata-rata:** O(n^2)
    *   **Kasus Terbaik:** O(n) (jika array sudah terurut)
*   **Kompleksitas Ruang:** O(1) (pengurutan in-place)

## 4. Merge Sort

### Pseudocode

```
FUNGSI merge_sort(array):
  n = panjang array
  JIKA n <= 1:
    KEMBALIKAN array

  mid = n / 2
  left_half = array[0 HINGGA mid-1]
  right_half = array[mid HINGGA n-1]

  left_half = merge_sort(left_half)
  right_half = merge_sort(right_half)

  KEMBALIKAN merge(left_half, right_half)

FUNGSI merge(left_array, right_array):
  result_array = daftar kosong
  i = 0, j = 0
  SELAMA i < panjang left_array DAN j < panjang right_array:
    JIKA left_array[i] <= right_array[j]:
      TAMBAHKAN left_array[i] KE result_array
      i = i + 1
    LAIN:
      TAMBAHKAN right_array[j] KE result_array
      j = j + 1
  SELAMA i < panjang left_array:
    TAMBAHKAN left_array[i] KE result_array
    i = i + 1
  SELAMA j < panjang right_array:
    TAMBAHKAN right_array[j] KE result_array
    j = j + 1
  KEMBALIKAN result_array
```

````python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    return merge(left_half, right_half)

def merge(left, right):
    result = []
    i = 0
    j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    while i < len(left):
        result.append(left[i])
        i += 1

    while j < len(right):
        result.append(right[j])
        j += 1

    return result

# Demo
example_list = [38, 27, 43, 3, 9, 82, 10]
print(f"Original list: {example_list}")
sorted_list = merge_sort(example_list)
print(f"Sorted list (Merge Sort): {sorted_list}")
````

### Analisis Kompleksitas

*   **Kompleksitas Waktu:**
    *   **Kasus Terburuk:** O(n log n)
    *   **Kasus Rata-rata:** O(n log n)
    *   **Kasus Terbaik:** O(n log n)
*   **Kompleksitas Ruang:** O(n) (karena kebutuhan array sementara selama penggabungan)
