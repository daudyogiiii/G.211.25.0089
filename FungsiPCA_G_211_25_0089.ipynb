```python
import numpy as np

def pca_from_scratch(X, n_components):
    # Langkah 1: Standarisasi data (pusatkan ke nilai rata-rata = 0)
    X_meaned = X - np.mean(X, axis=0)

    # Langkah 2: Hitung matriks kovarians
    # rowvar=False mengindikasikan bahwa kolom adalah fitur/variabel
    covariance_matrix = np.cov(X_meaned, rowvar=False)

    # Langkah 3: Hitung Nilai Eigen (Eigenvalues) dan Vektor Eigen (Eigenvectors)
    eigen_values, eigen_vectors = np.linalg.eigh(covariance_matrix)

    # Langkah 4: Urutkan Vektor Eigen berdasarkan Nilai Eigen terbesar ke terkecil
    sorted_index = np.argsort(eigen_values)[::-1]
    sorted_eigenvectors = eigen_vectors[:, sorted_index]

    # Langkah 5: Pilih sejumlah n_components teratas
    eigenvector_subset = sorted_eigenvectors[:, 0:n_components]

    # Langkah 6: Transformasikan data ke dimensi baru (Proyeksi Dot Product)
    X_reduced = np.dot(eigenvector_subset.transpose(), X_meaned.transpose()).transpose()

    return X_reduced

# --- Uji Coba Fungsi ---
X = np.array([[2.5, 2.4, 3.1],
              [0.5, 0.7, 1.2],
              [2.2, 2.9, 2.2],
              [1.9, 2.2, 4.0],
              [3.1, 3.0, 3.8]])

# Reduksi data dari 3 dimensi menjadi 2 dimensi
X_hasil = pca_from_scratch(X, n_components=2)

print("Hasil PCA From Scratch:\n", X_hasil)
```
