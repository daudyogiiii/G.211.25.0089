```python
def ListSum(arr, n):
    return sum(arr[:n])

arr = [5, 6, 1, 2]
n = len(arr)
print(ListSum(arr, n))
```

```python
n = 3
m = 3
nom = [(3, 2, 7), (2, 6, 8), (5, 1, 9)]
total_sum = 0

for i in range(n):
    for j in range(m):
        total_sum += nom[i][j]

print(f"{total_sum}")
```

```python
def factorial(n):
    """
    Menghitung nilai faktorial dari suatu bilangan n.
    Contoh: factorial(5) = 5 * 4 * 3 * 2 * 1 = 120
    """
    if not isinstance(n, int) or n < 0:
        raise ValueError("Input harus bilangan bulat non-negatif.")

    if n == 0:
        return 1

    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

number = 10
print(f"{number}! = {factorial(number)}")

number = 5
print(f"{number}! = {factorial(number)}")

number = 0
print(f"{number}! = {factorial(number)}")
```
